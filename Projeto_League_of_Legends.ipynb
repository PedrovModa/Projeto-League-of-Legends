## Importando bibliotecas

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

## Abrindo arquivo CSV

```python
dados = pd.read_csv('bans.csv')
```

```python
dados
```

## Excluindo colunas que não serão utilizadas no decorrer da análise

```python
dados.drop(columns = ['Address'], axis = 1, inplace = True)
dados.drop(columns = ['Team'], axis = 1, inplace = True)
```

## Excluindo valores nulos

```python
dados.dropna(how='all', axis = 1, inplace = True)
dados.dropna(how='any', axis = 0, inplace = True)
```

```python
dados
```

## Transformando todas as colunas em Series para posteriormente transformarmos em um DataFrame único

```python
ban_1 = dados['ban_1']
ban_2 = dados['ban_2']
ban_3 = dados['ban_3']
ban_4 = dados['ban_4']
ban_5 = dados['ban_5']
```

## Concatenando todas as Series

```python
dados_brutos = pd.concat([ban_1, ban_2, ban_3, ban_4, ban_5])
```

## Transformando-as em DataFrame

```python
dataset = pd.DataFrame(dados_brutos)
```

```python
dataset
```

## Renumerando o índex do meu DataFrame e renomeando o nome da minha coluna.

```python
dataset.index = range(dataset.shape[0])
```

```python
dataset.columns = ['Quantidade']
```

```python
dataset
```

## Verificando a quantidade de vezes que cada campeao foi banido e colocando essa informação em meu DF.

```python
quantidade = dataset['Quantidade'].value_counts()
```

```python
quantidade
```

```python
df_quantidade = pd.DataFrame(quantidade).reset_index()
```

```python
df_quantidade
```

## Renomeando e formatando meu índex para melhor visualização e também minha coluna de personagens.

```python
df_quantidade.rename(columns = {'index': 'Personagens'}, inplace = True)
```

```python
df_quantidade.columns.name = 'Colocação'
```

```python
df_quantidade.index = np.arange(1, len(df_quantidade) + 1)
```

```python
df_quantidade
```

## Criando variável que me permita visualizar a quantidade de números de partidas de acordo com a quantidade de bans.

```python
tamanho_df = dados.shape[0]
```

```python
tamanho_df
```

## Criando coluna porcentagem, para que se possa ter uma noção 

```python
df_quantidade['Porcentagem'] = ((df_quantidade['Quantidade']*100)/tamanho_df).round(4)
```
